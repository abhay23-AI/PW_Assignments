### 1. What is a random variable in probability theory?
A **random variable** is a numerical value that represents the outcome of a random process. It maps outcomes of a probabilistic experiment to real numbers.


### 2. What are the types of random variables?
- **Discrete Random Variable**: Takes countable values (e.g., number of heads in coin tosses).
- **Continuous Random Variable**: Takes an uncountable range of values (e.g., height, temperature).


### 3. What is the difference between discrete and continuous distributions?
- A **discrete distribution** applies to variables that take specific, separate values (e.g., rolling a die).
- A **continuous distribution** applies to variables that take an infinite number of values within a range (e.g., normal distribution of heights).


### 4. What are probability distribution functions (PDF)?
A **Probability Distribution Function (PDF)** defines the likelihood of each possible outcome of a random variable. 
- **For discrete variables**: Probability Mass Function (PMF)
- **For continuous variables**: Probability Density Function (PDF)


### 5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- The **PDF** gives the probability of a specific value.
- The **CDF** gives the probability that the variable is **less than or equal** to a given value.


### 6. What is a discrete uniform distribution?
A **discrete uniform distribution** assigns equal probability to all possible outcomes in a finite set (e.g., rolling a fair die).


### 7. What are the key properties of a Bernoulli distribution?
- Only two possible outcomes: success (1) or failure (0).
- Probability of success: $p$, probability of failure: $1 - p$.
- Mean: $p$, Variance: $p(1-p)$.

### 8. What is the binomial distribution, and how is it used in probability?
- Models the number of successes in $n$ independent Bernoulli trials.
- **Probability Mass Function (PMF)**:
  $$ P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k} $$
- Used in success/failure experiments like coin flips or quality control.



### 9. What is the Poisson distribution, and where is it applied?
- Models the probability of a given number of events happening in a fixed interval of time or space.
- **PMF**:
  $$ P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!} $$
- Applied in rare event counting (e.g., number of customer arrivals, radioactive decay).



### 10. What is a continuous uniform distribution?
- A distribution where all outcomes within a given interval are equally likely.
- **Probability Density Function (PDF)**:
  $$ f(x) = \frac{1}{b - a}, \quad a \leq x \leq b $$
- Used in simulations and random sampling.

### 11. What are the characteristics of a normal distribution?
- Symmetrical, bell-shaped curve.
- Defined by mean $\mu$ and standard deviation $\sigma$.
- **68-95-99.7% Rule (Empirical Rule)**:
  - 68% of data lies within 1$\sigma$.
  - 95% within 2$\sigma$.
  - 99.7% within 3$\sigma$.

### 12. What is the standard normal distribution, and why is it important?
- A normal distribution with mean $\mu = 0$ and standard deviation $\sigma = 1$.
- Used for Z-score calculations and statistical hypothesis testing.


### 13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- States that the **sampling distribution of the sample mean** approaches a normal distribution, regardless of the population’s distribution, if the sample size is large enough ($n \geq 30$).
- Important for inferential statistics and hypothesis testing.


### 14. How does the Central Limit Theorem relate to the normal distribution?
- It justifies using the normal distribution for inference, even when the original data is not normally distributed.


### 15. What is the application of Z statistics in hypothesis testing?
- Z-tests compare sample means to population means when the population standard deviation is known.
- Used in large samples ($n \geq 30$) for normal approximation.


### 16. How do you calculate a Z-score, and what does it represent?
- **Formula**:
  $$ Z = \frac{X - \mu}{\sigma} $$
- Measures how many standard deviations an observation is from the mean.


### 17. What are point estimates and interval estimates in statistics?
- **Point Estimate**: A single value estimate of a population parameter (e.g., sample mean).
- **Interval Estimate**: A range within which the population parameter is expected to fall, given a confidence level (e.g., confidence interval).


### 18. What is the significance of confidence intervals in statistical analysis?
- Provides a range of plausible values for a population parameter.
- Expresses uncertainty in estimates (e.g., “95% confidence interval”).


### 19. What is the relationship between a Z-score and a confidence interval?
- Z-score determines critical values for confidence intervals.
- **Confidence Interval Formula**:
  $$ \bar{X} \pm Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}} $$



### 20. How are Z-scores used to compare different distributions?
- Standardizes different distributions to a common scale (mean = 0, SD = 1).
- Allows comparison of scores from different normal distributions.


### 21. What are the assumptions for applying the Central Limit Theorem?
- Random sampling.
- Independent observations.
- Large sample size ($n \geq 30$) or normal population distribution.


### 22. What is the concept of expected value in a probability distribution?
- The long-run average outcome of a random variable.
- **Formula**:
  $$ E[X] = \sum x P(x) \quad \text{(discrete)}, \quad E[X] = \int x f(x) dx \quad \text{(continuous)} $$


### 23. How does a probability distribution relate to the expected outcome of a random variable?
- The probability distribution determines the likelihood of each outcome, and the expected value gives the weighted average of all possible outcomes.

### 02. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
def plot_discrete_uniform():
    x = np.arange(1, 7)
    pmf = stats.randint.pmf(x, 1, 7)
    plt.step(x, pmf, where='mid', color='blue')
    plt.xlabel("Outcome")
    plt.ylabel("Probability")
    plt.title("Discrete Uniform Distribution (Die Roll)")
    plt.show()

plot_discrete_uniform()
```

### 03. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution>

```python
def bernoulli_pdf(p, x):
    if not (0 <= p <= 1 and x in [0, 1]):
        raise ValueError("Invalid input: x should be 0 or 1, and p should be between 0 and 1.")
    return stats.bernoulli.pmf(x, p)

print("Bernoulli PDF:", bernoulli_pdf(0.5, 1))
```

### 04. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
def simulate_binomial():
    data = np.random.binomial(n=10, p=0.5, size=1000)
    plt.hist(data, bins=10, density=True, alpha=0.7, color='green')
    plt.xlabel("Successes")
    plt.ylabel("Frequency")
    plt.title("Binomial Distribution (n=10, p=0.5)")
    plt.show()

simulate_binomial()
```

### 05. Create a Poisson distribution and visualize it using Python.

```python
def poisson_distribution():
    x = np.arange(0, 15)
    pmf = stats.poisson.pmf(x, mu=5)
    plt.step(x, pmf, where='mid', color='red')
    plt.xlabel("Events")
    plt.ylabel("Probability")
    plt.title("Poisson Distribution (lambda=5)")
    plt.show()

poisson_distribution()
```

### 06. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.

```python
def plot_discrete_uniform_cdf():
    x = np.arange(1, 7)
    cdf = stats.randint.cdf(x, 1, 7)
    plt.step(x, cdf, where='mid', color='purple')
    plt.xlabel("Outcome")
    plt.ylabel("Cumulative Probability")
    plt.title("CDF of Discrete Uniform Distribution")
    plt.show()

plot_discrete_uniform_cdf()
```

### 09. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    return (data - mean) / std_dev

data = np.random.normal(50, 10, 1000)
z_scores = calculate_z_scores(data)
plt.hist(z_scores, bins=30, density=True, alpha=0.7, color='orange')
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.title("Z-score Distribution")
plt.show()
```

### 10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

```python
def central_limit_theorem():
    sample_means = np.mean(np.random.exponential(scale=2, size=(1000, 30)), axis=1)
    plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='green')
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.title("Central Limit Theorem in Action")
    plt.show()

central_limit_theorem()
```

### 20. Create a confidence interval for a dataset using Python and interpret the result

```python
import scipy.stats as stats

data = np.random.normal(100, 15, 50)
mean, sem = np.mean(data), stats.sem(data)
confidence_interval = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)

print(f"95% Confidence Interval: {confidence_interval}")
```

### 27. Implement Z-score normalization on a dataset and visualize its transformation

```python
def z_normalize(data):
    return (data - np.mean(data)) / np.std(data)

data = np.random.normal(100, 20, 1000)
normalized_data = z_normalize(data)

plt.hist(data, alpha=0.5, label="Original")
plt.hist(normalized_data, alpha=0.5, label="Normalized")
plt.legend()
plt.show()
```
